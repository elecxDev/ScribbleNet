```python
import os
import csv

ROOT_FOLDER = r"C:\Users\adrie\Documents\Projects\ScribbleNet\cvl-database-1-1"

OUTPUT_CSV = "cvl_words_dataset_detailed.csv"

rows = []

for root, dirs, files in os.walk(ROOT_FOLDER):

    if "words" not in root.lower():
        continue

    if any(skip in root.lower() for skip in ["lines", "pages", "xml"]):
        continue

    for file in files:
        if file.lower().endswith(".tif"):

            full_path = os.path.join(root, file)
            filename = file

            name_without_ext = os.path.splitext(file)[0]
            parts = name_without_ext.split("-")

            if len(parts) >= 5:
                writer_id = parts[0]
                text_id = parts[1]
                line_id = parts[2]
                word_index = parts[3]
                label = "-".join(parts[4:])  # handles hyphenated words
            else:
                continue

            rows.append([
                full_path,
                filename,
                writer_id,
                text_id,
                line_id,
                word_index,
                label
            ])

with open(OUTPUT_CSV, mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow([
        "full_path",
        "filename",
        "writer_id",
        "text_id",
        "line_id",
        "word_index",
        "label"
    ])
    writer.writerows(rows)

print(f"✅ Detailed CSV created with {len(rows)} samples!")
```

```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv("cvl_words_dataset_detailed.csv")

# Get unique writers
writers = df["writer_id"].unique()
np.random.shuffle(writers)

# Split writers
train_split = int(0.8 * len(writers))
val_split = int(0.9 * len(writers))

train_writers = writers[:train_split]
val_writers = writers[train_split:val_split]
test_writers = writers[val_split:]

# Create splits
train_df = df[df["writer_id"].isin(train_writers)]
val_df = df[df["writer_id"].isin(val_writers)]
test_df = df[df["writer_id"].isin(test_writers)]

# Save
train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
test_df.to_csv("test.csv", index=False)

print("✅ Train/Val/Test split created!")
print("Train samples:", len(train_df))
print("Val samples:", len(val_df))
print("Test samples:", len(test_df))
```
